# **JOINS**

# **FECHAS**

[Doc Fechas](https://docs.python.org/es/3/library/datetime.html#strftime-and-strptime-format-codes)

### **Trasnformar a Fehca la columna fecha_publicado.**

Nota: Crear una columna fecha_pulblicacion_norm

```python
# Las fechas por defecto son cadenas de texto
# Con Python podemos trasformarlas a fechas
# Poseer las fechas en el formato de Fechas de Python facilita la reestructuraciÃ³n de las mismas o la organizaciÃ³n del dataframe segÃºn esta columna.
df_final['fecha_pulblicacion_norm'] = pd.to_datetime(df_final['fecha_publicado'], format='%d/%m/%Y')
df_final.head()
```

### **Crear un columna con el dia de publicaciÃ³n**

```python
df_final['dia_semana'] = df_final['fecha_pulblicacion_norm'].apply(lambda x: x.strftime('%A'))
df_final.head()
```

```python
#Ejempo usando funciones
def darDiaSemana(fecha):
  diaSemana = fecha.strftime('%A')
  return diaSemana

df_final['dia_semana_fun'] = df_final['fecha_pulblicacion_norm'].apply(darDiaSemana)
df_final.head()
```

### **Crear un columna con el mes y aÃ±o de publicaciÃ³n**

```python
df_final['mes'] = df_final['fecha_pulblicacion_norm'].apply(lambda x: x.strftime('%m/%Y'))
df_final.head()
```

### **Sacar de la columna fecha_publicado el mes y aÃ±o**

```python
df_final['mes'] = df_final['fecha_publicado'].apply(lambda x: x[3:])
df_final.head()
```

```python
# Usando funcion
def extraer_mes_y_anio_de_cadena(fecha_str):
    # Dividir la cadena por el carÃ¡cter '/'
    partes = fecha_str.split('/')
    # El mes es el segundo elemento y el aÃ±o es el tercer elemento
    mes = partes[1]
    anio = partes[2]
    # Devolver el mes y el aÃ±o en el formato deseado
    return mes + '/' + anio
df_final['fecha_publicado'].apply(extraer_mes_y_anio_de_cadena)
df_final.head()
```

```python
# Funcion usando slice
def extraer_mes_y_anio_con_slice(fecha_str):
    # Extraer el mes y el aÃ±o usando rebanada de string
    mes = fecha_str[3:5]
    anio = fecha_str[6:10]
    # Devolver el mes y el aÃ±o en el formato deseado
    return mes + '/' + anio

df_final['fecha_publicado'].apply(extraer_mes_y_anio_con_slice)
df_final.head()
```
