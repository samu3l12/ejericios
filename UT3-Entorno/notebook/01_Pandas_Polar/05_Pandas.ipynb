# **Transformaci√≥n de Datos**
Apartir de los datos existentes, crear o calcular nuevascolumnas que tengan datos transformados.



#### Transformacion de datos de una columna
Ejem : Crear una columna llamada Ganancias, que va ser el numero de retweets de cada fila multipicado de un numero aleatoria del 3 al cinco.

```python
# Limpieza de Datos
df = df.fillna({"retweets" : 0,"mentions" : -1})

#Transformaci√≥n de datos

import random
def cacularGanancias(retweets):
  ganancia = retweets * random.randint(3,5)
  return ganancia

df["ganancias"] = df["retweets"].apply(cacularGanancias)

df.head(10)
```

#### Transformaci√≥n de datos de dos o mas columnas

Ejem: Crear la columna Popular, que es dividir cuantas personas sigo y cuantas personas me siguen


```python
#Transformacion de datos de dos columnas o mas

def calcularPopularidad(fila):
 resultado = fila["followees"]/fila["followers"]
 return resultado

df["popularidad"] = df.apply(calcularPopularidad, axis = 1) # Dado que estamos haciendo .apply() del DataFrame completo y no de una columna invidual, "axis = 1", le expecifica a metodo apply() que debe ser aplicado por cada fila del df.

df.head()

```

#### Sumar la informaci√≥n de otras columnas mediante la agregaci√≥n. Por lo tanto calcular el promedio de "Me gustas" en los tweet por p√°is.

```python
# AGRUPANDO Y AGREGANDO DATOS
# En recientes versiones de Pandas
# Tenemos que filtrar primero solo las columnas numericas y la columna por la que vamos a agrupar
df[
    ['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad']
].groupby("country").mean() # Agrupando por pais y sacando la media de cada columna

```

#### Agrupando por pais y aplicando diferentes funciones de agregacion a cada columna (followers: suma , mentions: media y retweets: maximo )

```python
# Agrupando por pais y aplicando diferentes funciones de agregacion a cada columna
df.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})

```

#### Crear un DataFrame llamado "grouped" que contenga la agrupaci√≥n anterior

```python
grouped = df.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})
```

#### Filtrar los follower de DataFrame anterior donde sean mayores de 5000

```python
grouped[grouped["followers"] > 5000]
```
